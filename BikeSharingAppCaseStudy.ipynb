### Data Preparation

```python
# List of variables to Remove as they does not help in anyway in this analysis
# Removing 'casual' and 'registered' columns as well as the objective is to 
# model the demand for shared bikes. 'casual' or 'registered' data  does not make sense

varlist =  ['instant', 'dteday', 'casual', 'registered']

# Droping varlist elements 
for element in varlist:

    bikesharing.drop(element, inplace=True, axis=1)

```

```python
# Check the BikeSharing dataframe now

bikesharing.head()
```

```python
# Converting season and weathersit variable to categorical string variable.  
bikesharing.season.replace({1:"spring", 2:"summer", 3:"fall", 4:"winter"},inplace = True)

bikesharing.weathersit.replace({1:'clear',2:'mist',3:'lightsnow',4:'heavyrain'},inplace = True)

bikesharing.head(600)
```

The variable `weekday` has 7 days.  converting weekdays to categorial strings like ('Mon', Tue',...). 


```python

bikesharing['weekday'].replace({0:"Sun",1:"Mon",2:"Tue",3:"Wed",4:"Thu",5:"Fri",6:"Sat"},inplace=True)
bikesharing['mnth'].replace({1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"},inplace=True)

bikesharing.head(10)
```

```python
# Since temp and atemp are higly corelated, we can drop one of them. Dropping Temp variable
bikesharing = bikesharing.drop(columns=['temp'])
bikesharing.head(100)
```

#Creating categorical variables to Dummy variable 


```python
cat_var_columns = ['mnth', 'season', 'weekday', 'weathersit']
status = pd.get_dummies(bikesharing[cat_var_columns],drop_first = True)
status.head(10)
```

```python
bikesharing = pd.concat([bikesharing, status], axis = 1)
```

```python
bikesharing.drop(cat_var_columns, axis = 1, inplace = True)
bikesharing.head(10)
```

```python
## Splitting the Data into Training and Testing Sets
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(bikesharing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features 

We will use min-max scaling.

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = [ 'atemp', 'hum', 'windspeed', 'cnt']

df_train[num_vars] = scale.fit_transform(df_train[num_vars])

df_train.head()
```

```python
df_train.describe()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (20, 30))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

## Building our model

we will be using the **LinearRegression function from SciKit Learn** for its compatibility with RFE (which is a utility from sklearn)

### RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

Results looks good. Can't find p value of any independent variable that can be dropped. The p value is less than  5% confidence level.

```python
#X_train_rfe = X_train_rfe.drop(['const'], axis=1)
#X_train_rfe.head()
X_train_rfe.columns
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_cnt = lm.predict(X_train_rfe)
```

```python
# Importing the required libraries for plots.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```
